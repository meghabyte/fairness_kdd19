```python
from matplotlib import pyplot as plt
import pandas as pd
results = pd.read_csv("Data/medical_logresults_t20.csv")
```

```python
ea = []
fd = []
fn = []
dp = []
for index, row in results.iterrows():
    if(type(row["Probability Tuple"]) is not str):
        continue
    ea.append((float(row["Probability Tuple"].split(',')[0][1:]), row["Gender"], row["Political"], row["Race"], row["Education"], row["Age"]))
    fd.append((float(row["Probability Tuple"].split(',')[1]), row["Gender"], row["Political"], row["Race"], row["Education"], row["Age"]))
    fn.append((float(row["Probability Tuple"].split(',')[2]), row["Gender"], row["Political"], row["Race"], row["Education"], row["Age"]))
    dp.append((float(row["Probability Tuple"].split(',')[3][:-1]), row["Gender"], row["Political"], row["Race"], row["Education"], row["Age"]))
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(15, 8))
counts, bins, patches = ax.hist([x[0] for x in ea], range=(0, 1), facecolor='orange', edgecolor='gray')
ax.set_xlabel("Likelihood of EP after 20 tests", fontsize="40")
ax.set_ylabel("% of participants", fontsize="40")
ax.set_ylim(0,80)
bin_centers = 0.5 * np.diff(bins) + bins[:-1]
counter = 0
locs = [325, 60, 80, 50, 60, 40, 30, 50, 40, 30]
strs = ["" if counts[i] == 0 else str(int(int(counts[i])*100.0/len([x[0] for x in ea])))+"%" for i in range(0,len(locs))]
for count, x in zip(counts, bin_centers):
    # Label the raw counts
    ax.annotate(strs[counter], xy=(x, 0), xycoords=('data', 'axes fraction'),
        xytext=(0, locs[counter]+30), textcoords='offset points', va='top', ha='center', fontsize="35")
    counter +=1 
plt.tick_params(axis='both', which='major', labelsize=35)
plt.tight_layout()
fig.savefig('Plots/med_ea.pdf')
plt.show()
```
