```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
```

```python
ec2 = []
randomtest = []
#These are text files containing the runs for the command python fairnessAppServer.py --testmode hypothesis=(0|1|2|3) depending on the hypothesis model
#The EC2 test selection and Random Test Selection can be toggled in the code - parameter control for this will be included in the future
for i in ["ea_ec_noisy.txt", "fd_ec_noisy.txt", "fn_ec_noisy.txt", "dp_ec_noisy.txt"]:
    f = open("Data/RandomTest/"+i, "r")
    ec2.append(f.readlines())
    f.close()
for i in ["accuracy_randomtest_noisy.txt", "fd_randomtest_noisy.txt", "fn_randomtest_noisy.txt", "dp_randomtest_noisy.txt"]:
    f = open("Data/RandomTest/"+i, "r")
    randomtest.append(f.readlines())
    f.close()
```

```python
hyps = ["Hypothesis Accuracy", "Hypothesis False Discovery", "Hypothesis False Negative", "Hypothesis Demographic Parity"]
ec2_probs = []
randomtest_probs = []
for i in range(0,len(ec2)):
    probs = []
    for line in ec2[i]:
        if(hyps[i] in line):
            probs.append(float(line.split()[-1]))
    ec2_probs.append(probs)
for i in range(0,len(randomtest)):
    probs = []
    for line in randomtest[i]:
        if(hyps[i] in line):
            probs.append(float(line.split()[-1]))
    randomtest_probs.append(probs)
```

```python
#EC2
fig, ax = plt.subplots(figsize=(15,8))
colors = ["orange","green","blue","red"]
for i in range(0, len(ec2_probs)):
    plt.plot(ec2_probs[i], color=colors[i], linewidth="2")
ax.set_xlabel("Number of tests administered", fontsize="30")
ax.set_ylabel("Likelihood", fontsize="30")
ax.set_xlim(0,150)
plt.tick_params(axis='both', which='major', labelsize=25)
red_patch = mpatches.Patch(color='red', label='DP')
orange_patch = mpatches.Patch(color='orange', label='EP')
blue_patch = mpatches.Patch(color='blue', label='FNP')
green_patch = mpatches.Patch(color='green', label='FDP')
plt.legend(handles=[red_patch, orange_patch, blue_patch, green_patch], fontsize=20, loc="right")
plt.savefig("Plots/EC2Tests_NoisyResponse.pdf")
plt.show()
```

```python
#Random Test
fig, ax = plt.subplots(figsize=(15,8))
colors = ["orange","green","blue","red"]
for i in range(0, len(randomtest_probs)):
    plt.plot(randomtest_probs[i], color=colors[i])
ax.set_xlabel("Number of tests administered", fontsize="30")
ax.set_ylabel("Likelihood", fontsize="30")
plt.tick_params(axis='both', which='major', labelsize=25)
red_patch = mpatches.Patch(color='red', label='DP')
orange_patch = mpatches.Patch(color='orange', label='EP')
blue_patch = mpatches.Patch(color='blue', label='FNP')
green_patch = mpatches.Patch(color='green', label='FDP')
plt.legend(handles=[red_patch, orange_patch, blue_patch, green_patch], fontsize=20, loc="right")
plt.savefig("Plots/RandomTests_NoisyResponse.pdf")
plt.show()
```
